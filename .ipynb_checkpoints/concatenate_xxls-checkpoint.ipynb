```python
import pandas as pd

#read file names
import os
def list_files_in_directory(directory_path):
    all_entries = os.listdir(directory_path)
    files = [entry for entry in all_entries if os.path.isfile(os.path.join(directory_path, entry))]
    xls_files = [file for file in files if '.xls' in file]
    return xls_files

#Check if a string represents a numeric value.
import re
def is_numeric_string(s):
    return bool(re.match(r'^\d+$', str(s)))
```

```python
#organize data between 2003 and 2009
#states = ['MA','SP','RJ','TO','PA','SE','MS','PB','RN','AL','GO','MT','RR','BA','CE','MG','PI','ES','RS','PR','PE','AP','SC']
df_2003_to_2009_and_2013_to_2015 = pd.DataFrame()
for year in ['2003','2004','2005','2006','2007','2008','2009','2013','2014','2015']:
    files = list_files_in_directory(str('data/'+year))
    for state in files:
        path = str('data/'+year+'/'+state)
        df = pd.read_excel(path)
        filtered_df = df[df.iloc[:, 0].apply(is_numeric_string)]
        df2 = filtered_df.copy()
        df2['ano'] = year
        df_2003_to_2009_and_2013_to_2015 = pd.concat([df_2003_to_2009_and_2013_to_2015, df2], ignore_index=True)

df_2003_to_2009_and_2013_to_2015#.head(5)
```

```python
#organize data between 2010 and 2012
#years 2010 and 2011 have no information about 'RITO'
df_2010_to_2012 = pd.DataFrame()
for year in ['2010','2011','2012']:
    files = list_files_in_directory(year)
    for state in files:
        path = str(year+'/'+state)
        df = pd.read_excel(path)
        filtered_df = df[df.iloc[:, 0].apply(is_numeric_string)]
        df2 = filtered_df.copy()
        df2['ano'] = year
        df2['UF'] = state.split('.')[0]
        df_2010_to_2012 = pd.concat([df_2010_to_2012, df2], ignore_index=True)


#give correct name to the columns
df_2010_to_2012.columns = ['Nº','Código IBGE ','Município','Nº do Decreto','Data do Decreto','Desastre','SE/ECP','Nº da Portaria','Data da Portaria','Nº do D.O.U.','Data do D.O.U.','ano','UF','Rito']
# Rearrange columns
cols = df_2010_to_2012.columns.tolist() # Convert column names to a list
cols.insert(13, cols.pop(cols.index('ano')))
cols.insert(1, cols.pop(cols.index('UF'))) # Move 'col3' to the second position
df_2010_to_2012 = df_2010_to_2012[cols] # Reindex DataFrame using the rearranged list


df_2010_to_2012.head(5)
```

```python
df_2016 = pd.DataFrame()
year = '2016'
files = list_files_in_directory(year)
for state in files:
    path = str(year+'/'+state)
    df = pd.read_excel(path)
    filtered_df = df[df.iloc[:, 1].apply(is_numeric_string)]
    df2 = filtered_df.copy()
    df2['ano'] = year
    df2 = df2.iloc[:,1:]
    df_2016 = pd.concat([df_2016, df2], ignore_index=True)

#df_2016 = df_2016.iloc[:,1:]
df_2016.columns = df_2003_to_2009_and_2013_to_2015.columns

df_2016#.head(5)
```

```python
#concatenate values of all years (2003 to 2016)
s2id = pd.concat([df_2003_to_2009_and_2013_to_2015, df_2010_to_2012,df_2016], ignore_index=True)
```

```python
#I created subclasses to bether agregate the dissasters 
classification = pd.read_csv('classification_desasters.csv')
s2id_ = s2id.merge(classification, left_on='Desastre', right_on='classe_3', how='left')

s2id_.head(5)
```

```python
import matplotlib.pyplot as plt


# Group by 'Year' and count the number of disasters
s2id_counts = s2id_.iloc[:,[13]+[14]].groupby('ano').size()

# Plotting
plt.figure(figsize=(10,6))
s2id_counts.plot(kind='line', marker='o')
plt.title('Frequency of Natural Disasters by Year')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
s2id_class1 = s2id_.groupby(['ano', 'classe_1']).size().unstack().fillna(0)
s2id_class1
```

```python
s2id_class = s2id_.groupby(['ano', 'classe_2']).size().unstack().fillna(0)
s2id_class
# Plotting
plt.figure(figsize=(10,6))
for disaster_type in s2id_class.columns:
    plt.plot(s2id_class.index, s2id_class[disaster_type], marker='o', label=disaster_type)

plt.title('Frequency of Natural Disasters by Year')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10,6))
for disaster_type in s2id_class.columns:
    plt.fill_between(s2id_class.index, s2id_class[disaster_type], label=disaster_type, alpha=0.5)

plt.title('Frequency of Natural Disasters by Year')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.legend(title='Disaster Type')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
disaster_counts = s2id_.groupby(['ano', 'classe_2']).size().unstack().fillna(0)

# Plotting
plt.figure(figsize=(10,6))
disaster_counts.plot.area(stacked=True, alpha=0.8, ax=plt.gca())

plt.title('Frequency of Natural Disasters by Year')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.legend(title='Disaster Type')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Year': [2000, 2000, 2001, 2001, 2001, 2002, 2002, 2002, 2002],
    'Disaster': ['Flood', 'Earthquake', 'Flood', 'Flood', 'Earthquake', 'Flood', 'Hurricane', 'Earthquake', 'Hurricane'],
    'Other_Info': [10, 20, 15, 18, 25, 12, 30, 22, 33]  # Just an example
}
df = pd.DataFrame(data)

# Group by 'Year' and 'Disaster' and then unstack to get separate columns for each disaster type
disaster_counts = df.groupby(['Year', 'Disaster']).size().unstack().fillna(0)

# Set a colormap
colormap = plt.cm.viridis

# Plotting
plt.figure(figsize=(10,6))
disaster_counts.plot.area(stacked=True, colormap=colormap, alpha=0.7, ax=plt.gca())

plt.title('Frequency of Natural Disasters by Year')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.legend(title='Disaster Type', loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=3)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
